```python
import re
```

```python
with open("miracle_in_the_andes.txt", "r", encoding="utf-8") as f:
    book = f.read()
```

```python
# Most used words that are not articles

patter = re.compile("[a-zA-Z]+")
findings = patter.findall(book.lower())
print(findings)
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
nltk.download('vader_lexicon')
```

```python
# Remove the stop words

english_stopwords = set(stopwords.words('english'))
findings = [word for word in findings if word not in english_stopwords]
```

```python
from collections import Counter

# Count the most common words
word_counter = Counter(findings)

# Get the 10 most common words
most_common_words = word_counter.most_common(10)

# Print the 10 most common words
print(most_common_words)
```

Sentiment Analysis: Most postive and negative chapter

```python
from nltk.sentiment import SentimentIntensityAnalyzer
```

```python
analyzer = SentimentIntensityAnalyzer()
```

```python
r = analyzer.polarity_scores("I hate you. I hate life.")
print(r)
```

```python
# Analyze chapter sentiment

pattern = re.compile("Chapter [0-9]+")
chapters = re.split(pattern, book)
chapters = chapters[1:]

for chapter in chapters:
    scores = analyzer.polarity_scores(chapter)
    print(scores)
```
